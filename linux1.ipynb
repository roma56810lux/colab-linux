```python
#@title Run
# Установка рабочего стола XFCE, VNC и необходимых компонентов
!echo "Пожалуйста подождите..."
!echo "deb [trusted=yes] https://packages.mozilla.org/apt mozilla main" | sudo tee -a /etc/apt/sources.list.d/mozilla.list > /dev/null
!sudo apt-get update > /dev/null
!printf "49\n1\n17\n" | apt-get install -y xfce4 xfce4-goodies tightvncserver novnc websockify firefox/mozilla > /dev/null

# Настройка пароля для VNC (по умолчанию "password", можно изменить в следующей строке)
!export USER=root 
vncpassword = "123456"
!printf "$vncpassword\n$vncpassword\nn\n" | vncpasswd

# Запуск VNC сервера
!sudo vncserver :1 -geometry 1280x800 -depth 24
!export DISPLAY=:1

# Запуск прокси (Websockify) для доступа через браузер
!websockify -D --web=/usr/share/novnc/ 6080 localhost:5901

# Установка и запуск LocalTunnel для создания публичного URL
!npm install -g localtunnel
import subprocess, requests
process = subprocess.Popen('lt --port 6080'.split(), stdout=subprocess.PIPE, stderr=subprocess.PIPE, bufsize=0)
output = process.stdout.readline().decode('utf-8').strip()+'/vnc.html'
url = output[output.find('https'):]

# Получение пароля для LocalTunnel
ltpassword = requests.get('https://loca.lt/mytunnelpassword').text

# Вывод ссылки и паролей в блокнот
from IPython import display
display.clear_output()
print(f'Настройка завершена!\n')
print(f'Откройте: {url}')
print(f'Введите: {ltpassword}')
print(f'Нажмите Click to Submit')
print(f'Нажмите Connect')
print(f'Введите: {vncpassword}')
print(f'Нажмите Send password')
```
